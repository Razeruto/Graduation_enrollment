```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
```

```python
# Load student data from CSV file
student_data = pd.read_csv('/content/student_data.csv')

# Display the first few rows of the data
student_data.head()
```

```python
# Split the data into training and testing sets
X = student_data.drop(['Enrollment_Status', 'Graduation_Success', 'Socioeconomic_Status'], axis=1)
y_enrollment = student_data['Enrollment_Status']
y_graduation = student_data['Graduation_Success']
X_train, X_test, y_enrollment_train, y_enrollment_test, y_graduation_train, y_graduation_test = train_test_split(X, y_enrollment, y_graduation, test_size=0.2, random_state=42)
```

```python
# Train a Random Forest classifier for enrollment prediction
enrollment_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
enrollment_classifier.fit(X_train, y_enrollment_train)

# Make predictions on the test set
enrollment_predictions = enrollment_classifier.predict(X_test)
```

```python
# Evaluate the enrollment classifier
print("Enrollment Classifier Accuracy:", accuracy_score(y_enrollment_test, enrollment_predictions))
print("Enrollment Classifier Report:\n", classification_report(y_enrollment_test, enrollment_predictions))
```

```python
# Train a Random Forest classifier for graduation prediction
graduation_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
graduation_classifier.fit(X_train, y_graduation_train)

# Make predictions on the test set
graduation_predictions = graduation_classifier.predict(X_test)
```

```python
# Evaluate the graduation classifier
print("Graduation Classifier Accuracy:", accuracy_score(y_graduation_test, graduation_predictions))
print("Graduation Classifier Report:\n", classification_report(y_graduation_test, graduation_predictions))
```
